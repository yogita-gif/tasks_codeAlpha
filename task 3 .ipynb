```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/car data.csv")
```

```python
df
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Selling_Price'], bins=20, kde=True)

plt.title("Distribution of Selling Price")
plt.xlabel("Selling Price (Lakhs)")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Fuel_Type', data=df)

plt.title("Cars by Fuel Type")
plt.show()
```

```python
plt.figure(figsize=(8,6))

numeric_df = df[['Year','Selling_Price','Present_Price','Driven_kms','Owner']]

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Factors Affecting Car Selling Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='Present_Price',
    y='Selling_Price',
    hue='Fuel_Type',
    data=df
)

plt.title("Present Price vs Selling Price")

plt.show()

corr = df['Present_Price'].corr(df['Selling_Price'])

print("Correlation:", round(corr,2))
print("INSIGHT:")
print("Higher present market value leads to higher resale value.")
print("Decision: Present Price should be the primary factor in pricing models.")
```

```python
top10 = df.sort_values(
    by='Selling_Price',
    ascending=False
).head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x='Selling_Price',
    y='Car_Name',
    data=top10
)

plt.title("Top 10 Highest Selling Price Cars")
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,8))

sns.histplot(df['Selling_Price'], ax=axes[0,0])
axes[0,0].set_title('Selling Price')

sns.countplot(x='Fuel_Type', data=df, ax=axes[0,1])
axes[0,1].set_title('Fuel Type')

sns.scatterplot(x='Present_Price',
                y='Selling_Price',
                data=df,
                ax=axes[1,0])
axes[1,0].set_title('Present vs Selling Price')

sns.scatterplot(x='Driven_kms',
                y='Selling_Price',
                data=df,
                ax=axes[1,1])
axes[1,1].set_title('Driven Kms vs Selling Price')

plt.tight_layout()
plt.show()
```
